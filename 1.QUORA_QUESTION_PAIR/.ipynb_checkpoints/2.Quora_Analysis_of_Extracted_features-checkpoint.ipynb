# ANALYSIS OF EXTRACTED WORD

## 1. word_share (How thw word share effect the target)

```python
plt.figure(figsize=(12, 8))
plt.subplot(1,2,1) #first subplot ie.Violin Plot
#plotting violin plot
sns.violinplot(x = 'is_duplicate', y = 'word_share', data = data2[0:])

#plotting the pdf 
plt.subplot(1,2,2) #second subplot ie.PDF 
sns.distplot(data2[data2['is_duplicate'] == 1.0]['word_share'][0:] , label = "1", color = 'red')
sns.distplot(data2[data2['is_duplicate'] == 0.0]['word_share'][0:] , label = "0" , color = 'blue' )
plt.show()
```
